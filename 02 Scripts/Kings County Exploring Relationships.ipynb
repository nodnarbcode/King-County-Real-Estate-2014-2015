### 2. Creating correlation heatmap 

```python
plt.matshow(df.corr())
plt.show()
```

```python
#Next, we add a legend, column labels, change the size of the heatmap and add a title
f=plt.figure(figsize=(15,15)) #This establishes the size of the heatmap
plt.matshow(df.corr(), fignum=f.number) #Defines the type of plot
plt.xticks(range(df.shape[1]), df.columns, fontsize=12, rotation=90)
plt.yticks(range(df.shape[1]), df.columns, fontsize=12) #No need for rotation in y-axis
cb=plt.colorbar() #add a color legend (optimizes itseld based on colors used in heatmap)
cb.ax.tick_params(labelsize=12) #add font size
plt.title('Correlation Matrix', fontsize=15) #Creates titles
```

#### Using the legend on the right, the darker the purple, the stronger the negative relationship is between the two variables. A lighter purple represents a no relationship between the two variables. The bright yellow represents a strong positive relationship.

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(15,15))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

#### The scatterplot helps identify exact values of correlation between two variables. It uses the same legend as the heatmap but gives more precise insight to any correlation.

#### There seems to be a strong positive correlation between bedrooms, bathrooms and square foot living. There is a strong negative correlation between year built and condition.

### 3. Creating a scatterplot

```python
# Create a scatterplot for the "price" and "sqft_living" columns in seaborn

sns.lmplot(x = 'price', y = 'sqft_living', data = df)
```

#### This scatterplot shows a strong positive correlation between price and sq_ft living. As price increases, the square foot of living also increases. 

```python
# Create a scatterplot for the "bathrooms" and "sqft_living" columns in seaborn

sns.lmplot(x = 'bathrooms', y = 'sqft_living', data = df)
```

#### This scatterplot shows a positive correlation between bathrooms and sqft_living. The more bathrooms, the higher the square footage of property. 

### 4. Pair plots

### 5. Categorical Plots

```python
#### Creating a categorical variable that splits the "price" column into categories
sns.histplot(df['price'], bins = 75, kde = True)
```

```python
# Creating new variables

df.loc[df['price'] < 500000, 'Price Level'] = 'Low price'
df.loc[(df['price'] >= 500000) & (df['price'] < 1200000), 'Price Level'] = 'Middle price'
df.loc[df['price'] >= 1200000, 'Price Level'] = 'High price'
```

```python
df['Price Level'].value_counts()
```

```python
# Create a categorical plot in seaborn using the price categories created above

sns.set(style="ticks")
g = sns.catplot(x="bedrooms", y="price", hue="Price Level", data=df)
```

#### The higher priced properties tend to have between 2 and 6 bedrooms.
